```python
from pyspark.sql.functions import *
df1=spark.read.table(silver_table)
df1.createOrReplaceTempView("silver")
df2=spark.sql("""select claim_status,insurance_provider,sum(claim_amount) sum_claim_amt,count(*) cnt 
              from telecom_silver.claims_silver 
              group by claim_status,insurance_provider 
              order by claim_status,insurance_provider""")
df2.show(20)
df2.write.mode("overwrite").saveAsTable(gold_fact_aggr_table)
```
