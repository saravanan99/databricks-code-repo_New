```python
# ==========================================
# 1. PATHS & TABLE CONFIGURATIONS
# ==========================================
# Table Names
silver_table = "telecom_silver.claims_silver"
```

```python
adlsacct='we47toolstorage'
containername='container1'
accesskey='replacewithproperaccesskey'
spark.conf.set(f"fs.azure.account.key.{adlsacct}.dfs.core.windows.net", accesskey)
path = f"abfss://{containername}@{adlsacct}.dfs.core.windows.net/we47adfdata/"
```

```python
from pyspark.sql.functions import *
df1=spark.read.parquet(path)
df2=df1.dropDuplicates().withColumn("insurance_provider",coalesce(col("insurance_provider"),lit("NA")))
df3=df2.where("claim_amount>1000")
df3.write.mode("overwrite").saveAsTable(silver_table)
```
